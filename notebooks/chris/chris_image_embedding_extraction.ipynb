```python
batch_size = 5000

input_file_path_image = "../raw_data/embeded_data/embedding_train_image.csv"
input_file_path_text = "../raw_data/embeded_data/embedding_train_text.csv"

output_file_path = "../raw_data/embeded_data/embedding_train.csv"

file_exists = os.path.isfile(output_file_path)

if file_exists:
    with open(output_file_path) as f:
        print(sum(1 for line in f))
        saved_batches = int(sum(1 for line in f)/batch_size)
        start_point = batch_size * saved_batches
else:
    start_point = 0

for i in range(start_point, 51301, batch_size):
    batch_image = pd.read_csv(input_file_path_image, skiprows=i, nrows=batch_size)
    batch_text = pd.read_csv(input_file_path_text, skiprows=i, nrows=batch_size)
    
    # Concatenate the batch DataFrames
    concatenated_batch = pd.concat([batch_image, batch_text], axis=1)
    
    # Append or overwrite the output file
    if file_exists:
        concatenated_batch.to_csv(output_file_path, mode='a', header=False, index=False)
    else:
        concatenated_batch.to_csv(output_file_path, mode='w', header=True, index=False)
        file_exists = True
```

```python
X_test = pd.read_csv("../raw_data/test_x.csv", delimiter=" ", index_col="Id")

# Define the image path
image_path = "../raw_data/Images/"

# Load the VGG16 model
vgg_model = VGG16(weights='imagenet', include_top=False)

# Preprocess and extract features for each image in X_test
output_features = np.zeros((len(X_test), 7, 7, 512))  # Initialize array for output features

for i, image_name in enumerate(X_test["Image_name"]):
    image = load_img(image_path + image_name)
    image = img_to_array(image)
    image = preprocess_input(image)
    image = np.expand_dims(image, axis=0)  # Add batch dimension
    
    output_features[i] = vgg_model.predict(image)  # Extract features using VGG16 model
    clear_session()

# Reshape the output features
output_features = output_features.reshape(len(X_test), -1)

# Save the output features to a CSV file
output_file_path = "../raw_data/embeded_data/embedding_test_image.csv"
pd.DataFrame(output_features).to_csv(output_file_path, header=True, index=False)
```
