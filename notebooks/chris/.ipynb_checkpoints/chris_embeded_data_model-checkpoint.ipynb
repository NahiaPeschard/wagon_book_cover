```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

from sklearn.model_selection import cross_validate
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, f1_score

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, Sequential
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import RMSprop, Adam, SGD
```

```python
def plot_history(history, title='', axs=None, exp_name=""):
    if axs is not None:
        ax1, ax2 = axs
    else:
        f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    if len(exp_name) > 0 and exp_name[0] != '_':
        exp_name = '_' + exp_name
    ax1.plot(history.history['loss'], label = 'train' + exp_name)
    ax1.plot(history.history['val_loss'], label = 'val' + exp_name)
    ax1.set_ylim(0., 5)
    ax1.set_title('loss')
    ax1.legend()

    ax2.plot(history.history['accuracy'], label='train accuracy'  + exp_name)
    ax2.plot(history.history['val_accuracy'], label='val accuracy'  + exp_name)
    ax2.set_ylim(0., 1.0)
    ax2.set_title('Accuracy')
    ax2.legend()
    return (ax1, ax2)
```

```python
def load_data(filename, batch_size):
    X_train_reader = pd.read_csv("../raw_data/embeded_data/embedding_train.csv", chunksize=batch_size, index_col=None)
    y_train_reader = pd.read_csv("../raw_data/train_y.csv", delimiter=" ", index_col="Id", chunksize=batch_size)
    data_reader = pd.read_csv(filename, chunksize=batch_size)
    for X_chunk, y_chunk in zip(X_train_reader, y_train_reader):
        X = X_chunk.values
        y = y_chunk['Genre_id'].values  
        yield X, to_categorical(y, num_classes=30)
```

```python
X_test = pd.read_csv("../raw_data/embeded_data/embedding_test.csv")
```

```python
y_test = pd.read_csv("../raw_data/test_y.csv", delimiter=" ", index_col="Id")['Genre_id']
```

```python
y_test_cat = to_categorical(y_test, num_classes=30)
```

```python
def initialize_model():
    model = Sequential()
    
    model.add(layers.Dense(320, activation="relu", input_dim=26889))
    
    model.add(layers.Dense(100, activation="relu"))
    
    model.add(layers.Dense(30, activation="softmax"))

    model.compile(loss="categorical_crossentropy", optimizer="rmsprop", metrics=["accuracy"])

    return model
```

```python
model = initialize_model()
model
```

```python
batch_size = 32
epochs = 100

es = EarlyStopping(patience=5, restore_best_weights=True, monitor="val_accuracy")

# Batch training loop
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")
    
    # Generate batches from training data
    data_generator = load_data('../raw_data/embeded_data/embedding_train.csv', batch_size)
    
    for X_batch, y_batch in data_generator:
        # Train the model on the current batch
        model.train_on_batch(X_batch, y_batch)
    
    # Evaluate the model on the testing set
    loss, accuracy = model.evaluate(X_test, y_test_cat)
    print(f"Validation loss: {loss:.4f}, Accuracy: {accuracy:.4f}")
    
    # Apply early stopping
    es.model = model
    es.best = 0
    if es.on_epoch_end(epoch, logs={'val_accuracy': accuracy}):
        print("Early stopping triggered.")
        break

        
        
```
